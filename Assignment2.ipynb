```python
import numpy as np

def create_matrix(n):
    return np.full((n, n), -1)

#print the matrix in a readable format
def print_matrix(matrix):
    print_matrix = np.where(matrix == -1, '_', matrix)  
    print("\nCurrent state of the board:")
    print(print_matrix)

#validate move
def valid(matrix, row, col):
    n = len(matrix)
    return 1 <= row <= n and 1 <= col <= n and matrix[row-1, col-1] == -1

def make_move(matrix, player, row, col):
    matrix[row-1, col-1] = player

#check if game has finished
def is_full(matrix):
    return np.all(matrix != -1)

def det(matrix):
    return np.linalg.det(matrix)

def matrix_game():
    n = int(input("Enter the size of the matrix (n): "))
    matrix = create_matrix(n)

    first_player = input("Who goes first, P0 or P1? ")
    player_A_turn = first_player == 'P0'


    while not is_full(matrix):
        print_matrix(matrix)
        if player_A_turn:
            print("P0's turn (0).")
        else:
            print("P1's turn (1).")

        try:
            row = int(input("Enter row number between 1 and n: "))
            col = int(input("Enter column number between 1 and n: "))
            if valid(matrix, row, col):
                make_move(matrix, 0 if player_A_turn else 1, row, col)
                player_A_turn = not player_A_turn  # Switch turns
            else:
                print("Invalid move. Try again.")
        except ValueError:
            print("Invalid input. Please enter numbers.")

    print_matrix(matrix)
    if det(matrix) == 0:
        print("P0 wins! (Determinant is 0)")
    else:
        print("P1 wins! (Determinant is not 0)")

matrix_game()
```
